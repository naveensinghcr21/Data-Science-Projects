```
#now capturing all the categorical columns(eg:bussiness travel,Department,etc)
categories = "object"

categorical_df = employee_df.select_dtypes(include=categories)
```

```
categorical_df
```

#need to convert these categorical values into numbers since our M.L model will not understand these values

#Skikitlearn provides one amazing tool called "ONE HOT ENCODER" from "sklearn.preprocessing" that encodes all such categorical values 

```
from sklearn.preprocessing import OneHotEncoder
onehotencoder= OneHotEncoder()
categorical_df = onehotencoder.fit_transform(categorical_df).toarray()
```

```
categorical_df
```

```
 categorical_df.shape
```

#now converting above "categorical_df" into dataframe

```
categorical_df = pd.DataFrame(categorical_df)
```

```
categorical_df
```

#now encoding numerical column values

```
numerics= "int64"
numerical_df= employee_df.select_dtypes(numerics) 
```

```
numerical_df.drop("Attrition",axis=1,inplace=True)
```

```
numerical_df
```

```
combined_df=pd.concat([categorical_df,numerical_df],axis=1)
```

```
combined_df
```
